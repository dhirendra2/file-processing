```python
import pandas as pd
import hashlib
import random
import string
import os, datetime
import uuid
```

```python
d0 = pd.read_csv('Sales.csv')
d0.head()
```

```python
d1 = d0[['Region', 'Country', 'Sales Channel', 'Order Date','Order ID', 'Total Revenue']].dropna()
display(d1.head())
```

```python
tmplist = list(set(d1['Country']))
```

```python
mapping1 = {i : (''.join(random.choice(string.hexdigits) for i in range(12)))+i+(''.join(random.choice(string.hexdigits) for i in range(12)))  for i in tmplist}
```

```python
d2 = d1.copy()
```

```python
d2['Hash'] = [mapping1[i] for i in d2['Country']]
display(d2.head())
```

```python
d2['NewHash'] = [hashlib.sha1(str.encode(str(i))).hexdigest() for i in d2['Hash']]
display(d2.head())
```

```python
d2.head()
```

```python
d2.drop('Date Time', inplace=True, axis=1)
```

```python
d2.head()
```

```python
d2.insert(6, column = "Date Time", value=datetime.datetime.now())
d2.insert(7, column = "ProcessIdentifier", value="")
```

```python
d2=d2.rename(columns={"NewHash": "RowHash"})
d2.head()
```
